```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Fancy plots
plt.style.use('fivethirtyeight')

# Data frame library 
import pandas as pd
```

```python
schoolpolutiondata = pd.read_csv('school polution.csv')
schoolpolutiondata
```

```python
primary_data = schoolpolutiondata[schoolpolutiondata['Phase of education'] == 'Primary']
```

PERCTOT = percentage of absence 

```python
Absences = pd.read_csv('england_abs.csv')
```

```python
Absences_polution=pd.merge(Absences, primary_data, on='URN')
Absences_polution = Absences_polution.drop (['Unnamed: 12', 'Close Date', 'PPERSABS15'], axis=1)
```

Absences

```python
Absences_polution.plot.scatter('PERCTOT', 'NO2ug/m3 mean 2013', title = 'Absences-Polution')
```

```python
np.median(Absences_polution['PERCTOT'])
```

```python
#Here I split the table in Above (institutions above the polution limit at the time) and Under (institutions below the polution limit at the time)
Above = Absences_polution.loc[Absences_polution['Above limit']=='Yes']
Under = Absences_polution.loc[Absences_polution['Above limit']=='No']
```

```python
np.median(Above['PERCTOT'])
```

```python
np.median(Under['PERCTOT'])
```
